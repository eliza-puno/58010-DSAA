```python
def quicksort(tasks):
    if len(tasks) <= 1:
        return tasks
    else:
        pivot = tasks.pop()

        less_than_pivot = []
        greater_than_pivot = []

        for task in tasks:
            if task['deadline'] < pivot['deadline']:
                less_than_pivot.append(task)
            else:
                greater_than_pivot.append(task)

        return quicksort(less_than_pivot) + [pivot] + quicksort(greater_than_pivot)

def manage_tasks():
    tasks = []

    while True:
        print("\nTask Management Menu:")
        print("1. Add a Task")
        print("2. View Tasks")
        print("3. Quit")

        choice = input("Enter your choice (1/2/3): ")

        if choice == "1":
            task_name = input("Enter task name: ")
            deadline = input("Enter task deadline (format: YYYY-MM-DD): ")

            task = {"name": task_name, "deadline": deadline}
            tasks.append(task)

            print(f"Task '{task_name}' added successfully!")

        elif choice == "2":
            if not tasks:
                print("No tasks to display.")
            else:
                sorted_tasks = quicksort(tasks)
                print("\nTasks sorted by deadline:")
                for idx, task in enumerate(sorted_tasks, start=1):
                    print(f"{idx}. {task['name']} - Deadline: {task['deadline']}")

        elif choice == "3":
            print("Exiting Task Management Program. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

def manage_budget():
    budget = float(input("Enter your monthly budget: "))
    expenses = []

    while True:
        print("\nBudget Management Menu:")
        print("1. Add an Expense")
        print("2. View Expenses")
        print("3. Quit")

        choice = input("Enter your choice (1/2/3): ")

        if choice == "1":
            expense_name = input("Enter expense name: ")
            expense_amount = float(input("Enter expense amount: "))

            expense = {"name": expense_name, "amount": expense_amount}
            expenses.append(expense)

            print(f"Expense '{expense_name}' added successfully!")

        elif choice == "2":
            if not expenses:
                print("No expenses to display.")
            else:
                total_expenses = sum(expense['amount'] for expense in expenses)
                remaining_budget = budget - total_expenses

                print("\nExpenses:")
                for idx, expense in enumerate(expenses, start=1):
                    print(f"{idx}. {expense['name']} - Amount: {expense['amount']}")

                print(f"\nTotal Expenses: {total_expenses}")
                print(f"Remaining Budget: {remaining_budget}")

        elif choice == "3":
            print("Exiting Budget Management Program. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

if __name__ == "__main__":
    while True:
        print("\nMain Menu:")
        print("1. Task Management")
        print("2. Budget Management")
        print("3. Quit")

        main_choice = input("Enter your choice (1/2/3): ")

        if main_choice == "1":
            manage_tasks()
        elif main_choice == "2":
            manage_budget()
        elif main_choice == "3":
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")
```
